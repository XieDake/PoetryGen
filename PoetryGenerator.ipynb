```python
import re
import jieba
import jieba.analyse


poetrys = []
poetry = ''
for line in open("poetry"):
    if line.find('。') != -1:
        # 方法1:先通过分隔符 '},' 来分隔字符串,在判断 'author' 并输出 author对应的内容
        str = re.split(r'，|？|。', line)
        poetry = (poetry + str[0] + ',' + str[1] + ',')
    else:
        if len(poetry) == 32:
            poetrys.append(poetry)
        if len(poetry) == 64:
            poetrys.append(poetry[0:31])
            poetrys.append(poetry[32:63])
        poetry = ''


        
train_triples = []
for poetry in poetrys:
    lines = poetry.split(',')
    for idx in range(len(lines)):
        key = ''
        train_triple = {}
        seg_list = jieba.cut(lines[idx], cut_all=False)
        print(lines[idx])
        for x, w in jieba.analyse.textrank(lines[idx], 1, withWeight=True):
            key = x
        train_triple['key'] = key
        train_triple['pretext'] = ';'.join(lines[:idx] if idx > 0 else '')
        train_triple['curtext'] = lines[idx]
        print(train_triple)
        train_triples.append(train_triple)
```
